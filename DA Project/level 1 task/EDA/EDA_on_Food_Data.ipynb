```python
plt.figure(figsize=(10, 6))
category_avg_calories = df.groupby('Category')['Calories'].mean().sort_values()
category_avg_calories.plot(kind='barh', color='Purple')
plt.title('Average Calories per Category')
plt.xlabel('Calories')
plt.ylabel('Category')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Total Fat', y='Carbohydrates', hue='Category')
plt.title('Total Fat vs Carbohydrates by Category')
plt.xlabel('Total Fat (g)')
plt.ylabel('Carbohydrates (g)')
plt.legend(loc='upper right')
plt.show()
```

```python
# Select only the numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix on the numeric data
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(14, 8))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='magma', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap Matrix')
plt.show()
```

```python
high_cal = df[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)
print("\n===== TOP 10 HIGH CALORIE ITEMS =====")
print(high_cal)

high_cal.plot(x='Item', y='Calories', kind='bar', figsize=(10,4))
plt.title("Top 10 High-Calorie Items")
plt.xlabel("Item")
plt.ylabel("Calories")
plt.show()
```

```python
print("\n===== LOWEST CALORIE ITEMS =====")
print(df[['Item', 'Calories']].sort_values(by='Calories').head(10))
```
